```python
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
from skimage import io
from sklearn.cluster import KMeans
from matplotlib.pyplot import imshow, imsave
%matplotlib inline
```

```python
face = misc.face()
```

```python
imshow(face)
```

```python
face.shape
```

```python
face_reshaped = (face / 255.0).reshape(-1,3)
```

```python
kmeans = KMeans(n_clusters=5).fit(face_reshaped)
compressed_face=kmeans.cluster_centers_[kmeans.labels_]
compressed_face=np.reshape(compressed_face, (face.shape))
```

```python
compressed_face.shape
```

```python
imsave('compressed_face.png',compressed_face)
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
img=mpimg.imread('compressed_face.png')
imgplot = plt.imshow(img)
plt.show()
imshow(face)
```
